1.	Machine Learning (Pembelajaran Mesin) adalah cabang dari kecerdasan buatan (Artificial Intelligence) yang memungkinkan sistem komputer untuk belajar dan meningkatkan kinerja mereka dalam tugas-tugas tertentu tanpa harus diprogram secara eksplisit. Proses ini melibatkan penggunaan algoritma dan model statistik untuk menganalisis dan mengidentifikasi pola dalam data. 3.	Berikut adalah beberapa macam taxonomi yang sering digunakan dalam machine learning:

a.	Berdasarkan jenis pembelajaran:
-	Supervised Learning
adalah metode di mana model dilatih menggunakan data yang telah dilabeli. Artinya, setiap contoh data yang digunakan dalam pelatihan sudah memiliki jawaban atau hasil yang benar. Model ini belajar dari data tersebut untuk memprediksi label dari data baru yang tidak dilabeli. Contoh umum dari supervised learning termasuk algoritma seperti Linear Regression, Logistic Regression, Support Vector Machines (SVM), Decision Trees, Random Forests, dan Neural Networks. Supervised learning banyak digunakan dalam aplikasi seperti klasifikasi email spam, di mana sistem perlu menentukan apakah sebuah email adalah spam atau bukan, dan prediksi harga rumah berdasarkan fitur-fitur tertentu seperti lokasi dan ukuran.



-	Unsupervised Learning
berbeda karena model dilatih menggunakan data yang tidak memiliki label. Model ini mencari pola atau struktur tersembunyi dalam data. Algoritma yang umum digunakan termasuk K-Means Clustering, Hierarchical Clustering, Principal Component Analysis (PCA), dan Apriori Algorithm. Aplikasi unsupervised learning termasuk klasterisasi pelanggan berdasarkan pola pembelian mereka atau menemukan hubungan antar produk yang sering dibeli bersama dalam analisis keranjang pasar.

-	Supervised Learning
adalah metode di mana agen belajar untuk membuat keputusan dengan berinteraksi dengan lingkungan. Agen menerima reward atau punishment berdasarkan tindakan yang diambil, dan tujuan agen adalah memaksimalkan reward kumulatif. Algoritma seperti Q-Learning, Deep Q-Networks (DQN), dan Proximal Policy Optimization (PPO) sering digunakan. Penerapan reinforcement learning meliputi permainan komputer seperti AI yang dapat bermain catur, serta navigasi robotik di mana robot belajar untuk bergerak dalam lingkungan tertntu untuk mencapai tujuan.

b.	Berdasarkan jenis output:
-	Klasifikasi (Classification)
Model membuat prediksi dengan output berupa kategori diskrit. Contohnya adalah mengidentifikasi apakah email adalah spam atau tidak, atau mengklasifikasikan jenis bunga berdasarkan fitur-fiturnya.

-	Regresi (Regression)
Digunakan ketika output yang diharapkan adalah nilai kontinu. Contohnya termasuk memprediksi harga rumah berdasarkan fitur-fitur tertentu atau memperkirakan permintaan produk di masa depan.

-	Klasterisasi (Clustering)
Adalah teknik di mana model mengelompokkan data yang tidak dilabeli menjadi beberapa kelompok atau klaster berdasarkan kesamaan tertentu. Contohnya adalah mengelompokkan pelanggan berdasarkan pola pembelian mereka atau segmentasi pasar.

-	Pengurangan Dimensi (Dimensionality Reduction)
Digunakan untuk mengurangi jumlah fitur dalam dataset, mempermudah visualisasi, dan meningkatkan efisiensi algoritma pembelajaran mesin. Principal Component Analysis (PCA) adalah contoh teknik pengurangan dimensi yang sering digunakan.

c.	Berdasarkan teknik yang digunakan:
-	Algoritma Linear
Seperti Linear Regression dan Logistic Regression mengasumsikan bahwa hubungan antara input dan output adalah linear. Algoritma ini sederhana dan cepat, namun mungkin tidak cukup untuk menangani hubungan yang kompleks dalam data.

-	Algoritma Non-Linear
Seperti Decision Trees, Random Forests, Support Vector Machines (SVM) dengan kernel, dan Neural Networks dapat menangani hubungan non-linear antara input dan output. Algoritma ini lebih fleksibel dan dapat menghasilkan prediksi yang lebih akurat dalam banyak kasus.

-	Algoritma Berbasis Probabilistik
Seperti Naive Bayes, Hidden Markov Models (HMM), dan Bayesian Networks menggunakan model probabilistik untuk membuat prediksi. Algoritma ini sering digunakan dalam aplikasi yang memerlukan pemodelan ketidakpastian.

-	Algoritma Berbasis Instance
Seperti K-Nearest Neighbors (KNN) dan Case-Based Reasoning membuat prediksi berdasarkan instance yang paling mirip dalam data latih. Algoritma ini sederhana namun dapat sangat efektif dalam beberapa kasus.

d.	Berdasarkan arsitektur model
-	Model Linear
Seperti Linear Regression dan Logistic Regression mengasumsikan hubungan linear antara input dan output. Model ini sederhana dan mudah dipahami, tetapi mungkin tidak cukup kuat untuk menangani hubungan yang kompleks.

-	Model Non-Linear
Seperti Decision Trees dan Neural Networks dapat menangani hubungan non-linear antara input dan output. Model ini lebih fleksibel dan dapat digunakan untuk berbagai jenis masalah yang lebih kompleks.

-	Ensemble Methods
Seperti Random Forests, Gradient Boosting Machines (GBM), dan AdaBoost menggabungkan beberapa model untuk meningkatkan kinerja prediksi. Teknik ini sering menghasilkan hasil yang lebih baik dibandingkan model tunggal karena memanfaatkan kekuatan dari berbagai model.

e.	Berdasarkan tujuan akhir
-	Prediksi
Prediksi adalah tujuan utama dalam banyak aplikasi machine learning, di mana model digunakan untuk memprediksi nilai masa depan berdasarkan data historis. Contohnya termasuk prediksi harga saham, cuaca, atau permintaan produk.

-	Deteksi Anomali (Anomaly Detection)
Deteksi Anomali (Anomaly Detection) digunakan untuk mengidentifikasi data yang tidak biasa atau mencurigakan. Contohnya termasuk mendeteksi transaksi penipuan dalam sistem keuangan atau mendeteksi kerusakan mesin dalam pemeliharaan prediktif.

-	Generative Models
Seperti Generative Adversarial Networks (GANs) dan Variational Autoencoders (VAEs) digunakan untuk membuat data baru yang mirip dengan data pelatihan. Teknik ini berguna dalam aplikasi seperti penciptaan gambar sintetis atau teks sintetis.

```python
# library
import pandas as pd

# Dataset
data = {
    'Hari': ['Senin', 'Selasa', 'Rabu', 'Kamis', "Jum'at", 'Sabtu', 'Minggu'],
    'Datang': [2, 3, 4, 1, 2, 5, 2],
    'Biaya': [30000, 35000, 25000, 15000, 20000, 30000, 35000],
    'Mahasiswa': ['Ani', 'Budi', 'Jono', 'Lono', 'Joni', 'Ani', 'Budi']
}

df = pd.DataFrame(data)

# a.
average_daily_visits = df['Datang'].mean()
print(f"a. Rata-rata mahasiswa datang per hari: {average_daily_visits:.2f}")

# b.
df['Total_Biaya'] = df['Datang'] * df['Biaya']
highest_cost_day = df.loc[df['Total_Biaya'].idxmax()]['Hari']
highest_cost = df['Total_Biaya'].max()
print(f" ")
print(f"b. Biaya tertinggi terjadi pada hari {highest_cost_day} sebesar Rp {highest_cost}")

# c.
days_high_cost = df[df['Total_Biaya'] > 110000]['Hari'].tolist()
print(f" ")
print(f"c. Hari dengan biaya lebih dari Rp 110,000: {days_high_cost}")

# d.
student_visits = df.groupby('Mahasiswa')['Datang'].sum()
most_frequent_student = student_visits.idxmax()
most_frequent_visits = student_visits.max()
print(f" ")
print(f"d. Mahasiswa yang paling banyak datang ke kampus: {most_frequent_student} dengan {most_frequent_visits} kedatangan")

# e.
students_sunday = df[df['Hari'] == 'Minggu']['Mahasiswa'].tolist()
print(f" ")
print(f"e. Mahasiswa yang datang pada hari Minggu: {students_sunday}")

# f.
highest_daily_cost = df['Total_Biaya'].max()
lowest_daily_cost = df['Total_Biaya'].min()
print(f" ")
print(f"f. ")
print(f"Biaya harian tertinggi: Rp {highest_daily_cost}")
print(f"Biaya harian terendah: Rp {lowest_daily_cost}")

# g.
highest_visit_frequency = df['Datang'].max()
lowest_visit_frequency = df['Datang'].min()
print(f" ")
print(f"g. ")
print(f"Frekuensi kedatangan tertinggi: {highest_visit_frequency} kali")
print(f"Frekuensi kedatangan terendah: {lowest_visit_frequency} kali")

# Menampilkan dataframe
df
```
